```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv("titanic_train.csv")
df_test = pd.read_csv("titanic_test.csv")
```

```python
df_train["Age"].fillna(df_train["Age"].median(),inplace=True)
df_test["Age"].fillna(df_test["Age"].median(),inplace=True)
```

```python
df_train["Cabin"]=df_train["Cabin"].replace(np.nan,"N")
df_test["Cabin"]=df_test["Cabin"].replace(np.nan,"N")
```

```python
df_test["Fare"].fillna(df_test["Fare"].median(),inplace=True)
```

```python
df_train["Embarked"]=df_train["Embarked"].replace(np.nan,"U")
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
df_train.drop(["Name","Ticket","Cabin"],axis = 1,inplace = True)
```

```python
df_train.head()
```

```python
df_test.drop(["Name","Ticket","Cabin"],axis = 1,inplace = True)
df_test.head()
```

```python
df_train.describe()
```

```python
for col in df_train.columns:
    if(df_train[col].dtypes == "O"):
        print("unique is {} = {}".format(col,df_train[col].unique()))
```

```python
for col in df_test.columns:
    if(df_test[col].dtypes == "O"):
        print("unique is {} = {}".format(col,df_test[col].unique()))
```

```python
dict = {
    'S':0,
    'C':1,
    'Q':2
}
```

```python
dict1 = {
    'S':0,
    'C':1,
    'Q':2,
    'U':3
}
```

```python
df_train["Embarked"]=df_train["Embarked"].map(dict1)
df_test["Embarked"]=df_test["Embarked"].map(dict)
```

```python
df_train = pd.get_dummies(df_train,drop_first=True)
df_test = pd.get_dummies(df_test,drop_first=True)
```

```python
df_test.head()
```

```python
df_test.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_train[["Age","Fare"]] = sc.fit_transform(df_train[["Age","Fare"]])
df_test[["Age","Fare"]] = sc.fit_transform(df_test[["Age","Fare"]])
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train.drop("PassengerId",axis = 1,inplace = True)
df_test.drop("PassengerId",axis = 1,inplace = True)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
# model bulding
x_train=df_train.iloc[:,1:].values
y_train=df_train.iloc[:,0].values
x_test=df_test.iloc[:,:].values
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='newton-cg',multi_class='ovr')
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
output = pd.DataFrame({'Gender': "gender", 'Survived': y_pred})
output.to_csv('Output.csv', index=False)
```
